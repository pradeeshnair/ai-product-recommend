```python
#Creating heat map to find correlation between features

import seaborn as sns

corr = df.corr()
top_corr_features = corr.index
plot.figure(figsize=(20,20))
#plot heat map
g=sns.heatmap(df[top_corr_features].corr(),annot=True,cmap="RdYlGn")
```

```python
# Remove one feature from data set whose correlation is high compared to the other

columns = np.full((corr.shape[0],), True, dtype=bool)
#print(columns)
for i in range(corr.shape[0]):
    for j in range(i+1, corr.shape[0]):
        if corr.iloc[i,j] >= 0.85:
            if columns[j]:
                columns[j] = False
#print(corr.shape[0])
#print(df.columns)
selected_columns = df.columns[columns]
df = df[selected_columns]
```

```python
#Rename churnid to target
df.rename(columns={'churnid':'target'},inplace=True)
```

```python
#Building classification tree model
```

```python
from sklearn.model_selection import train_test_split
from sklearn import tree
import itertools
from sklearn.metrics import confusion_matrix
```

```python
X = df.drop("target", axis=1)#['sepal length (cm)']

# Select output variable
y = df["target"]

```

```python
X.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y,test_size=0.3, random_state=0)
```

```python
classification_tree = tree.DecisionTreeClassifier(class_weight=None, criterion='gini', max_depth=None,
            max_features=3, max_leaf_nodes=None, min_samples_leaf=1,
            min_samples_split=5, min_weight_fraction_leaf=0.0,
            presort=False, random_state=None, splitter='best')

classification_tree = classification_tree.fit(X_train,y_train )
```

```python
y_pred = classification_tree.predict(X_test)
y_pred
```

```python
def plot_confusion_matrix(cm, classes,
                          normalize=False,
                          title='Confusion matrix',
                          cmap=plot.cm.Blues):
  
    print('Confusion matrix, without normalization')

    print(cm)

    plot.imshow(cm, interpolation='nearest', cmap=cmap)
    plot.title(title)
    plot.colorbar()
    tick_marks = np.arange(len(classes))
    plot.xticks(tick_marks, classes, rotation=45)
    plot.yticks(tick_marks, classes)

    fmt = '.2f' if normalize else 'd'
    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, format(cm[i, j], fmt),
                 horizontalalignment="center",
                 color="white" if cm[i, j] > thresh else "black")

    plot.ylabel('True label')
    plot.xlabel('Predicted label')
    plot.tight_layout()
    
# Compute confusion matrix
cnf_matrix = confusion_matrix(y_test, y_pred)
np.set_printoptions(precision=2)

# Plot non-normalized confusion matrix
plot.figure()
plot_confusion_matrix(cnf_matrix, classes=['0','1'],
                      title='Confusion matrix, without normalization')


plot.show()
```
